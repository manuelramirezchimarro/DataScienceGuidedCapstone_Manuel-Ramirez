```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('/Users/manuelramirez/Documents/diabetes presentation/diabetes_012_health_indicators_BRFSS2015.csv')
# Create a binary diabetes column: 1 = Diabetes, 0 = No Diabetes/Prediabetes
df['Diabetes_binary'] = df['Diabetes_012'].apply(lambda x: 1 if x == 2 else 0)
# 1. Diabetes Prevalence Bar Chart
plt.figure(figsize=(6,4))
diabetes_counts = df['Diabetes_binary'].value_counts().sort_index()
sns.barplot(x=['No Diabetes', 'Diabetes'], y=diabetes_counts.values, palette='pastel')
plt.title('Diabetes Prevalence')
plt.ylabel('Number of Individuals')
plt.tight_layout()
plt.savefig('diabetes_prevalence.png', dpi=300)
plt.show()

# Set the style for seaborn
sns.set(style='whitegrid')
```

```python
# 2. BMI vs Diabetes Boxplot
plt.figure(figsize=(6,4))
sns.boxplot(x='Diabetes_binary', y='BMI', data=df, palette='Set2')
plt.xticks([0,1], ['No Diabetes', 'Diabetes'])
plt.title('BMI Distribution by Diabetes Status')
plt.tight_layout()
plt.savefig('bmi_vs_diabetes.png', dpi=300)
plt.show()
```

```python
# 3. Age Distribution by Diabetes Histogram
plt.figure(figsize=(6,4))
sns.histplot(data=df, x='Age', hue='Diabetes_binary', multiple='stack', bins=20, palette='Set1')
plt.title('Age Distribution by Diabetes Status')
plt.xlabel('Age Category')
plt.tight_layout()
plt.savefig('age_vs_diabetes.png', dpi=300)
plt.show()
```

```python
# 4. Health Indicators Correlation Heatmap
plt.figure(figsize=(12,10))
corr = df.corr()
sns.heatmap(corr, cmap='coolwarm', center=0)
plt.title('Correlation Matrix of Health Indicators')
plt.tight_layout()
plt.savefig('correlation_matrix.png', dpi=300)
plt.show()
```

```python
# 5. Smoking vs Diabetes Bar Chart
plt.figure(figsize=(6,4))
smoking_diabetes = df.groupby('Diabetes_binary')['Smoker'].mean()
sns.barplot(x=['No Diabetes', 'Diabetes'], y=smoking_diabetes.values, palette='Blues')
plt.title('Smoking Rate by Diabetes Status')
plt.ylabel('Proportion of Smokers')
plt.tight_layout()
plt.savefig('smoking_diabetes.png', dpi=300)
plt.show()
```

```python
# 6. Alcohol Drinking vs Diabetes Bar Chart
plt.figure(figsize=(6,4))
alcohol_diabetes = df.groupby('Diabetes_binary')['HvyAlcoholConsump'].mean()
sns.barplot(x=['No Diabetes', 'Diabetes'], y=alcohol_diabetes.values, palette='Purples')
plt.title('Heavy Alcohol Consumption Rate by Diabetes Status')
plt.ylabel('Proportion of Heavy Alcohol Consumers')
plt.tight_layout()
plt.savefig('alcohol_diabetes.png', dpi=300)
plt.show()
```
